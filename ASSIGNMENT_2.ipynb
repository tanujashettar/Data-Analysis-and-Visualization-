Synthetic Dataset

```python
import pandas as pd

data = {
    "Emp_ID": [101,102,103,104,105,106,107,108],
    "Department": ["IT","HR","Finance","IT","Marketing","HR","Finance","IT"],
    "Work_Hours": [8.5,7.0,9.0,6.5,8.0,7.5,9.2,6.8],
    "Attendance": [95,85,98,80,90,88,96,75],
    "Performance_Score": [88,72,92,65,85,78,94,60]
}

df = pd.DataFrame(data)
df
```

EDA & Analysis

```python
# Summary statistics
df.describe()

# Department-wise performance
dept_avg = df.groupby("Department")["Performance_Score"].mean()

# Correlation
correlation = df[["Work_Hours","Attendance","Performance_Score"]].corr()

# Top & low performers
top3 = df.nlargest(3, "Performance_Score")
bottom3 = df.nsmallest(3, "Performance_Score")

dept_avg, correlation, top3, bottom3
```

Visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram
plt.hist(df["Performance_Score"])
plt.title("Performance Score Distribution")
plt.show()

# Bar chart
dept_avg.plot(kind="bar")
plt.title("Department-wise Average Performance")
plt.show()

# Heatmap
sns.heatmap(correlation, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
